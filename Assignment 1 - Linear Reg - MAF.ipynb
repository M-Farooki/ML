```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
#Relationship between Year and Per capita income
plt.scatter(df["year"], df["per capita income (US$)"], color = "lightcoral")
plt.title("Year vs Per Capita Income")
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.box(False)
plt.show()
```

```python
# Splitting variables
X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:]  # dependent
```

```python
X
```

```python
y
```

```python
train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
print(regressor.intercept_)
print(regressor.coef_)
print('y = ', regressor.intercept_, '+', regressor.coef_, '* X')
```

```python
# evaluate the model
print('MSE = ', mean_squared_error(y_test, y_pred_test))
print('R2 = ', r2_score(y_test, y_pred_test))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred_test)))
```

```python
# plot the model and data
plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred_train, marker='o', color='red')
plt.title('Year vs Per Capita Income')
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.legend(['X_train/y_train','X_train/Pred(y_test)'], title = 'Yr/Capita Income', loc='best', facecolor='white')
plt.show()
```
